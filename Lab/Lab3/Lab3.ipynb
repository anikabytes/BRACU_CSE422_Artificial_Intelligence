```python
import random

id = input()
for_random = []
id = list(id)
random_list = []

def prune(alpha,beta) :
    if (alpha >= beta) :
        return True

def mini(current_state,pointer,alpha,beta) :

    if (current_state != 0) :
        for child in range(0,2) :
            send_to_maxim= maxim(current_state-1,(pointer*2)+child,alpha,beta)
            beta = min(random_list[pointer],beta)
            if (prune(alpha,beta)) :
                break
            else :
                return send_to_maxim

    return random_list[current_state]

def maxim(current_state,pointer,alpha,beta) :

    if (current_state == 0) :
        return random_list[current_state]
    else :
        for child in range(0,2) :
            send_to_mini = mini(current_state-1,(pointer*2)+child,alpha,beta)
            alpha = max(random_list[pointer],alpha)
            if (prune(alpha,beta)) :
                break
            else :
                return send_to_mini


for i in range(0,len(id)) :
    if (id[i] == "0") :
        id[i] = "8"
    for_random.append(int(id[i]))

winning_point = int("".join(id[::-1][:2:]))
find_max = round(winning_point*1.5)
find_min = int(id[3])

winner = ""
leaf = 3
initial_alpha = -99999
initial_beta = 99999
root = 0

for i in range(0,len(for_random)) :
    random_list.append(random.randint(find_min,find_max))
alphaBeta_results = maxim(leaf,root,initial_alpha,initial_beta)
print(f"Generated 8 random points between the minimum and maximum point limits:{random_list}\nTotal points to win: {winning_point}")

if (alphaBeta_results >= winning_point) :
    winner = "Optimus Prime"
else :
    winner = "Megatron"
print(f"Achieved points by alpha-beta pruning = {alphaBeta_results}\nThe winner is {winner}\n")
random_list.clear()


# Task 2
shuffle = int(id[3])
all_points = []
def shuffle_store_list(shuffle_times) :

    if shuffle_times != 0 :
        for j in range(0,len(for_random)) :
            random_list.append(random.randint(find_min,find_max))
        all_points.append(maxim(leaf,root,initial_alpha,initial_beta))
        random_list.clear()
        shuffle_store_list(shuffle_times-1)

shuffle_store_list(int(id[3]))
print(f"After shuffle:\nList of all point values from each shuffle:{all_points}\nThe maximum value of all shuffles: {max(all_points)}")
winning_count = 0
for acquired in all_points :
    if (acquired >= winning_point) :
        winning_count += 1
print(f"Won {winning_count} times out of {shuffle} number of shuffles ")
```
