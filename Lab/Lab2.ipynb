```python
import random
batsman_info = {}
targett = 0
generation_num = []
file_path = "/content/drive/MyDrive/CSE422/Lab2/21101298_Lab2_input.txt"

def printResult(final) :

    print(list(batsman_info.keys()))
    if final == -1 :
        print(final)
    else :
        sequence = ""
        for values in batsman_info.values() :
            if values in final :
                sequence += "1"
            else:
                sequence += "0"
        print(sequence)
    read_file.close()


def fitnessFunc(select) :
    fit = []
    summ = [targett]

    checker = False
    pick_target = -1

    for j in select :
        take = sum(j)
        summ.append(take)
        if take == targett :
            for k in range(0,len(j)-1) :
                if (j[k] != j[k+1]) :
                    checker = True
                    pick_target = j


    select_min_index = summ.index(min(summ))
    if select_min_index < len(summ) - 1 :
        select.pop(select_min_index)
    return select, pick_target, checker

def mutationFunc(c1,c2) :

    c1[random.randint(0,len(c1)-1)] = random.choice(bats_info)
    c2[random.randint(0,len(c2)-1)] = random.choice(bats_info)

    return c1,c2

def crossoverFunc(parent1,parent2) :

    child1 = parent1[:len(parent1)-2] + parent2[len(parent2)-1:]
    child2 = parent2[len(parent2)-2:] + parent1[len(parent1)-1:]
    after_mutation_child1, after_mutation_child2 = mutationFunc(child1,child2)
    return after_mutation_child1, after_mutation_child2


def selection(parents) :
    generation_num.append(True)
    selected = []
    check = []
    iterate = 0
    upto = 20

    while iterate < upto :
        parent_leng = len(parents)
        parent1 = random.choice(parents)
        parent2 = random.choice(parents)
        if (parent1 != parent2) :
            c1,c2 = crossoverFunc(parent1,parent2)
            if (c1 not in selected) and (c2 not in selected) :
                if (c1 != c2) :
                    selected.append(c1)
                    selected.append(c2)
                else:
                    selected.append(c1)
        iterate += 1

    fit_population, target, flag = fitnessFunc(selected)

    if (len(generation_num) != 10) and (flag == False):
        selection(fit_population)
    else :
        printResult(target)

def initialPopulation(bats) :
    generation_num.append(True)
    check = 1
    initial = []
    iterate = 0
    upto = 20

    while iterate < upto :
        main = []
        for i in range(0,len(bats)) :
            gene = random.choice(bats)
            if gene not in main :
                main.append(gene)

        initial.append(main)
        iterate += 1
    fit_population, target, flag = fitnessFunc(initial)
    selection(fit_population)


batsman_info_file = open(file_path,"a")

number = int(input())
target = int(input())

if (1 <= number <= 18) and (1 <= target <= 1000) :
    batsman_info_file.write(str(number)+" "+str(target)+"\n")
    for info in range(0,number) :
        name = input()
        runs = int(input())
        if (1 <= runs<= 401):
                batsman_info_file.write(name+" "+str(runs)+"\n")
batsman_info_file.close()

read_file = open(file_path,"r")
info_file = read_file.readlines()
targett = 0

for info in range(0,len(info_file)) :
    info_file[info] = info_file[info].strip("\n").split(" ")
    if info == 0 :
        targett = int(info_file[info][1])
    else:
        batsman_info[info_file[info][0]] = int(info_file[info][1])
bats_info = list(batsman_info.values())
initialPopulation(bats_info)
```
